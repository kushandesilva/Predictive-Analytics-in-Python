```python
#Random sampling â€“ splitting a dataset in training and testing datasets
import pandas as pd
data = pd.read_csv('Customer Churn Model.txt')
len(data) 
```

```python
#generate random numbers and create a filter on which to partition the data:
a=np.random.randn(len(data))
check=a<0.8
training=data[check]
testing=data[~check]
# rows where the value of the random number is less than 0.8 becomes a part of the training variable
#while the one with a value greater than 0.8 becomes a part of the testing dataset
```

```python
len(testing) 
```
